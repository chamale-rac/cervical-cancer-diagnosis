## [5] Categorical Variable Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Asumiendo que df es tu DataFrame
numerical_columns = [
    "Age",
    "Number of sexual partners",
    "First sexual intercourse",
    "Num of pregnancies",
    "Smokes (years)",
    "Smokes (packs/year)",
    "Hormonal Contraceptives (years)",
    "IUD (years)",
    "STDs (number)",
    "STDs: Number of diagnosis",
]

# Identificar columnas categóricas
categorical_columns = list(set(df.columns) - set(numerical_columns) - set(["Dx"]))
categorical_data = df[categorical_columns]

# Crear gráficos de pastel para todas las variables categóricas
num_plots = len(categorical_columns)
num_cols = 2
num_rows = (num_plots + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(14, num_rows * 6))

for i, col in enumerate(categorical_columns):
    ax = axes[i // num_cols, i % num_cols]
    categorical_data[col].value_counts().plot(
        kind="pie",
        autopct=lambda p: f"{p:.1f}% ({int(p * sum(categorical_data[col].value_counts()) / 100)})",
        ax=ax,
        colors=sns.color_palette("viridis", len(categorical_data[col].unique())),
    )
    ax.set_title(f"Pie chart of {col}")
    ax.set_ylabel("")

# Ajustar diseño para que no se solapen los gráficos
plt.tight_layout()
plt.show()

# Tablas de contingencia y test chi-cuadrado
chi2_results = {}
for col in categorical_columns:
    contingency_table = pd.crosstab(df[col], df["Dx"])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    chi2_results[col] = p

chi2_results = pd.Series(chi2_results).sort_values()
print("P-values of Chi-Square Test:\n", chi2_results)

# Gráficos de barras apiladas para variables significativas
significant_columns = chi2_results[chi2_results < 0.05].index
for col in significant_columns:
    contingency_table = pd.crosstab(df[col], df["Dx"])
    contingency_table.plot(
        kind="bar", stacked=True, figsize=(10, 6), colormap="viridis"
    )
    plt.title(f"Distribution of {col} by Dx")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Heatmaps de proporción para variables significativas
for col in significant_columns:
    contingency_table = pd.crosstab(df[col], df["Dx"], normalize="index")
    plt.figure(figsize=(10, 6))
    sns.heatmap(contingency_table, annot=True, cmap="viridis")
    plt.title(f"Proportion of Dx by {col}")
    plt.show()
```

### Comments
There's a lot to unpack here, so let's break it down:

- We performed Chi-Square tests to identify the categorical variables that are significantly associated with the target variable `Dx`, revealing strong associations for `Dx:Cancer`, `Dx:HPV`, `Dx:CIN`, `Biopsy`, and `IUD`.

- Visualized the distribution of significant categorical variables using stacked bar charts and proportion heatmaps, highlighting the differing distributions of `Dx` across these variables.

- Found that variables like `Hormonal Contraceptives`, `STDs`, and several specific STDs did not show significant associations with `Dx`, suggesting they may not be strong predictors in this context.

## [7] Categorical variables transformations verification

```python
# Assuming df_cleaned is your DataFrame with the required data
data = df_cleaned.copy()

# List of continuous variables
continuous_vars = ['Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies',
                   'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives (years)', 'IUD (years)']

# List of binary variables
binary_vars = ['Smokes', 'Hormonal Contraceptives', 'IUD', 'STDs', 'STDs:condylomatosis',
               'STDs:vaginal condylomatosis', 'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis',
               'STDs:pelvic inflammatory disease', 'STDs:genital herpes', 'STDs:molluscum contagiosum',
               'STDs:HIV', 'STDs:Hepatitis B', 'STDs:HPV', 'Dx:Cancer', 'Dx:CIN', 'Dx:HPV', 'Hinselmann',
               'Schiller', 'Citology', 'Biopsy']

# List of multi-class variables to one-hot encode
multi_class_vars = ['STDs (number)', 'STDs: Number of diagnosis']

# Standardize continuous variables
scaler = StandardScaler()
data[continuous_vars] = scaler.fit_transform(data[continuous_vars])

# Standardize binary variables
data[binary_vars] = scaler.fit_transform(data[binary_vars])

# One-hot encode multi-class variables
data = pd.get_dummies(data, columns=multi_class_vars)

# Print unique values of Dx to confirm the number of classes
print("Unique values in Dx:", data['Dx'].unique())

# Perform PCA without multi-class categorical variables
data_no_multi_class = df_cleaned.copy()
data_no_multi_class[continuous_vars] = scaler.fit_transform(data_no_multi_class[continuous_vars])
data_no_multi_class[binary_vars] = scaler.fit_transform(data_no_multi_class[binary_vars])

X_no_multi_class = data_no_multi_class.drop(columns=['Dx'])
pca_no_multi_class = PCA(n_components=3)
pca_result_no_multi_class = pca_no_multi_class.fit_transform(X_no_multi_class)
total_var_no_multi_class = pca_no_multi_class.explained_variance_ratio_.sum() * 100

print(f"Total explained variance without multi-class categorical variables: {total_var_no_multi_class:.2f}%")

# Perform PCA with all variables
X_with_all = data.drop(columns=['Dx'])
pca_with_all = PCA(n_components=3)
pca_result_with_all = pca_with_all.fit_transform(X_with_all)
total_var_with_all = pca_with_all.explained_variance_ratio_.sum() * 100

print(f"Total explained variance with all variables: {total_var_with_all:.2f}%")

# Calculate cumulative explained variance for both cases
def calculate_cumulative_variance(data, n_components=30):
    X = data.drop(columns=['Dx'])
    pca = PCA(n_components=n_components)
    pca_result = pca.fit_transform(X)
    explained_variance = pca.explained_variance_ratio_
    cumulative_explained_variance = np.cumsum(explained_variance) * 100
    return cumulative_explained_variance

cumulative_var_no_multi_class = calculate_cumulative_variance(data_no_multi_class)
cumulative_var_with_all = calculate_cumulative_variance(data)

# Plot cumulative explained variance for both cases
def plot_cumulative_variance(cumulative_var, title=''):
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(cumulative_var) + 1), cumulative_var, marker='o', linestyle='--')
    plt.xlabel('Number of Principal Components')
    plt.ylabel('Cumulative Explained Variance (%)')
    plt.title(f'Cumulative Explained Variance by PCA Components\n{title}')
    plt.grid(True)
    plt.show()

plot_cumulative_variance(cumulative_var_no_multi_class, title='Without multi-class categorical variables')
plot_cumulative_variance(cumulative_var_with_all, title='With all variables')
```

### Comments

Based on the analysis, including multi-class categorical variables in PCA does not significantly improve the explained variance for the first few principal components, as indicated by the marginal decrease from 29.91% to 29.48%. This suggests that the additional information from these variables does not enhance the variance captured by PCA substantially. Furthermore, scaling binary variables, which transforms them to have a mean of 0 and variance of 1, ensures they contribute equally to the analysis. However, this transformation does not introduce substantial new information to the PCA when multi-class variables are included. For the specific purposes of this analysis, it may not be worth the effort to include multi-class categorical variables due to the added preprocessing complexity and the minimal impact on explained variance.
